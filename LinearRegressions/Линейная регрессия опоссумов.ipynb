```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import mean_squared_error
```

Сперва без категориальных переменных

```python
data1 = pd.read_csv('/content/sample_data/possum.csv')
data1
```

```python
data1 = data1.drop(['case', 'site', 'Pop', 'sex'], axis=1)
data1.dropna(inplace=True)
data1
```

```python
X = data1.drop(['totlngth'], axis=1)
y = data1['totlngth']
X = (X - X.mean(axis=0)) / (X.std(axis=0))
X_constant = sm.add_constant(X)
X_train, X_test, y_train, y_test = train_test_split(X_constant, y, train_size=0.7, test_size=0.3, random_state = 42)
```

```python
results1 = sm.OLS(y_train, X_train).fit()
results1.summary()
```

Посчитаем VIF

```python
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif['variable'] = X_train.columns
vif
```

Значение 1 указывает на отсутствие корреляции между данной независимой переменной и любыми другими независимыми переменными в модели.   

Значение от 1 до 5 указывает на умеренную корреляцию между данной объясняющей переменной и другими независимыми переменными в модели, но часто она недостаточно серьезна, чтобы требовать внимания.

Видим, что коэфициент детерминации выше 0.7, чем ближе значение коэффициента к 1, тем сильнее зависимость.
Коэфициент кореляции это корень из R^2 и он равен 0.85965109201, что говорит о сильной связи.

```python
predict_train = results1.predict(X_train)
predict_test = results1.predict(X_test)

mse_train = mean_squared_error(y_train, predict_train)
mse_test = mean_squared_error(y_test, predict_test)

print(f'MSE на тренировочном наборе: {mse_train:.4f}')
print(f'MSE на тестовом наборе: {mse_test:.4f}')
```

Теперь попробуем с категориальными данными

```python
data2 = pd.read_csv('/content/sample_data/possum.csv')
data2
```

```python
data2.sex = data2.sex.map({'m': 1, 'f': 0})
data2.Pop = data2.Pop.map({'Vic': 1, 'other': 0})
data2 = data2.join(pd.get_dummies(data2.site, prefix='site'))
data2.drop(['site', 'case'], axis=1, inplace=True )
data2.dropna(inplace=True)
```

```python
data2
```

```python
X = data2.drop(['totlngth'], axis=1)
y = data2['totlngth']
X = (X - X.mean(axis=0)) / (X.std(axis=0))
X_constant = sm.add_constant(X)
X_train, X_test, y_train, y_test = train_test_split(X_constant, y, train_size=0.7, test_size=0.3, random_state = 42)
```

```python
results2 = sm.OLS(y_train, X_train).fit()
results2.summary()
```

В данном случае, видим выше коэфициент детерминации равные 0.828, коэфициент корреляции 0.90994505328, что говорит о сильной связи.


```python
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif['variable'] = X_train.columns
vif
```

Тут мы видим громадные значения, (inf) это обычно указывает на сильную линейную зависимость между предикторами, есть мультиколлинеарность.

```python
predict_train = results2.predict(X_train)
predict_test = results2.predict(X_test)

mse_train = mean_squared_error(y_train, predict_train)
mse_test = mean_squared_error(y_test, predict_test)

print(f'MSE на тренировочном наборе: {mse_train:.4f}')
print(f'MSE на тестовом наборе: {mse_test:.4f}')
```

Т.К чем MSE ближе к нулю тем лучше, делаем вывод что во втором случае с категориальными признаками, лучше чем без.
Но во втором случае видим, что VIf показывает мультиколлинеарность, что не есть хорошо.
